```python
import os
import csv
import websocket
import json
import time
import threading
import joblib

import numpy as np
from collections import deque
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score

from src.websocket_client import run_websocket_in_thread, start_websocket, stop_websocket, clear_data
from src import websocket_client
```

```python
MODEL_DIR = "model"
```

```python
scaler = joblib.load(os.path.join(MODEL_DIR,'scaler.pkl'))
model = load_model(os.path.join(MODEL_DIR,'model.keras'))
```

```python
import random
random_list = [random.randint(0, 9) for _ in range(20)]
random_list
```

```python
def predict_next(scaler, model, input):
    input = np.array(input)
    scaled_input = scaler.transform(input.reshape(-1,1))
    prediction = model.predict(scaled_input.reshape(1,-1), verbose=0)
    rescaled_value = scaler.inverse_transform(prediction)
    result = np.abs(np.round(rescaled_value))[0, 0]
    return int(result)
```

```python
predict_next(scaler, model, random_list)
```

```python
timepoint_list = deque(maxlen=20)
ldp_list = deque(maxlen=20)
stop_event = threading.Event()
```

```python
def generate_sequence():
    stop_event.clear()
    
    # Fetch historical data
    start_websocket('history', 20)
    hist_data = websocket_client.fetched_data
    hist_time = hist_data[-1]['history']['times']
    hist_price = hist_data[-1]['history']['prices']
    hist_ldps = [int("{:.2f}".format(value)[-1]) for value in hist_price]
    timepoint_list.extend(hist_time)
    ldp_list.extend(hist_ldps)

    # Fetch live data
    live_thread = threading.Thread(target=start_websocket, args=('live',))
    live_thread.start()

    def store_live_data():
        while not stop_event.is_set():
            if websocket_client.fetched_data:
                live_data = websocket_client.fetched_data[0]
                live_time = live_data['tick']['epoch']
                live_price = live_data['tick']['quote']
                live_ldps = int("{:.2f}".format(live_price)[-1])
                timepoint_list.append(live_time)
                ldp_list.append(live_ldps)
                print(live_time)
            time.sleep(2)
        print("Data fetching stopped.")

    data_thread = threading.Thread(target=store_live_data)
    data_thread.start()

    return live_thread, data_thread
```

```python
stop_event = threading.Event()

def stop_threads():
    stop_event.set()
    print("Stopping threads...")
    stop_websocket()
```

```python
live_thread, data_thread = generate_sequence()
```

```python
stop_threads()
```
